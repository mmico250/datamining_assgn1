```python
import random
import time
import matplotlib.pyplot as plt


def random_trials(n, m):
    results = []
    for i in range(m):
        s = set()
        k = 0
        while len(s) < n:
            s.add(random.randint(0, n-1))
            k += 1
        results.append(k)
    return results

def plot_density(results, n, m):
    plt.hist(results, bins=50, density=True)
    plt.title(f"Density Plot for n={n}, m={m}")
    plt.xlabel("k")
    plt.ylabel("Density")
    plt.show()

def plot_run_time(n_list, m_list):
    times = []
    for n in n_list:
        start = time.time()
        random_trials(n, m_list[0])
        end = time.time()
        times.append(end-start)
    plt.plot(n_list, times)
    plt.title("Run Time Plot")
    plt.xlabel("n)")
    plt.ylabel("Run Time (s)")
    plt.show()

n = 10000
m = 1000
results = random_trials(n, m)
plot_density(results, n, m)

n_list = [100, 200, 300, 400, 500]
m_list = [100, 200, 300, 400, 500]
plot_run_time(n_list, m_list)
```

```python
import random
import time
import matplotlib.pyplot as plt

def random_trials(n, m):
    results = []
    for i in range(m):
        s = set()
        k = 0
        while True:
            r = random.randint(0, n-1)
            k += 1
            if r in s:
                break
            s.add(r)
        results.append(k)
    return results

def plot_density(results, n, m):
    plt.hist(results, bins=50, density=True)
    plt.title(f"Density Plot for n={n}, m={m}")
    plt.xlabel("Number of trials (k)")
    plt.ylabel("Density")
    plt.show()

def plot_run_time(n_list, m_list):
    times = []
    for n in n_list:
        start = time.time()
        random_trials(n, m_list[0])
        end = time.time()
        times.append(end-start)
    plt.plot(n_list, times)
    plt.title("Run Time Plot as n changes")
    plt.xlabel("Size of domain (n)")
    plt.ylabel("Run Time (s)")
    plt.show()
    times = []
    for m in m_list:
        start = time.time()
        random_trials(n_list[0], m)
        end = time.time()
        times.append(end-start)
    plt.plot(m_list, times)
    plt.title("Run Time Plot as m changes")
    plt.xlabel("Size of domain (m)")
    plt.ylabel("Run Time (s)")

n = 5000
m = 300
results = random_trials(n, m)
plot_density(results, n, m)

n_list = [1000, 2000, 3000, 4000, 5000]
m_list = [100, 200, 300, 400, 500]

plot_run_time(n_list, m_list)
```

```python
# k is the size of the shingle
def shingle(text: str, k: int):
    shingle_set = []
    for i in range(len(text) - k+1):
        shingle_set.append(text[i:i+k])
        # use the set function to eliminate dupes
    return set(shingle_set)
```

```python
def get_two_words(text: str): 
    two_word_gram = []
    new_a=text
    new_a = text.replace("."," ")
    split=new_a.split(" ")

    for i in range(0,len(split)-1,2):
        if (split[i]+" "+split[i+1]) not in two_word_gram:
            two_word_gram.append(split[i]+" "+split[i+1])

    return set(two_word_gram)

print("number of 2 word grams in very short sentences ",len(get_two_words(a)))
print("number of 2 word grams in short sentences ",len(get_two_words(b)))
print("number of 2 word grams in long short sentences ",len(get_two_words(c)))
print("number of 2 word grams in spanish short sentences ",len(get_two_words(d)))
```
